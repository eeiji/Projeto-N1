```python
from fastapi import FastAPI
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from pydantic import  BaseModel
```

```python
app = FastAPI()

# Dados simulados (ex: incidentes)
data = {
    "tickets_abertos": [10, 50, 30, 80, 20, 60, 15],
    "tempo_resolucao": [2, 10, 5, 15, 3, 12, 2],
    "risco": [0, 1, 0, 1, 0, 1, 0] # 1 = risco alto
}

df = pd.DataFrame(data)
```

```python
# Treinando modelo simples
X = df[["tickets_abertos", "tempo_resolucao"]]
y = df["risco"]

model = RandomForestClassifier()
model.fit(X, y)
```

```python
# Modelo de Entrada

class DadosEntrada(BaseModel):
    tickets_abertos: int
    tempo_resolucao: float #em horas
```

```python
# Endpoint da API
@app.post("/avaliar-risco")
def avaliar_risco(dados: DadosEntrada):

    entrada = [[dados.tickets_abertos, dados.tempo_resolucao]]
    pred = model.predict(entrada)[0]

    risco = "ALTO" if pred == 1 else "BAIXO"
    
    return {
        "tickets": dados.tickets_abertos,
        "tempo": dados.tempo_resolucao,
        "risco_previsto": risco
    }
```

```python
# Gráfico 

tempo_resolucao = [2, 3, 5, 2, 8, 10, 3, 6, 7, 5, 4, 6]

plt.hist(tempo_resolucao, bins= 8, edgecolor = 'black')
plt.title ("Distribuição de Tempo de Resolução")
plt.xlabel("Tempo de Resolução (Horas)")
plt.ylabel("Frequência")
plt.show()
```
